1. Calculating SSxx.

```python
xmean = x.mean()
xmean
```

```python
df['diffx'] = xmean - x
```

```python
df['diffx_square'] =df.diffx**2
```

```python
df
```

```python
SSxx = df.diffx_square.sum()
```

2. Define SSyy

```python
ymean = y.mean()
ymean
```

```python
df['diffy']= ymean - y
df
```

3. X-Y

```python
SSxy = (df.diffx * df.diffy).sum()
```

```python
m= SSxy/SSxx
m
```

```python
b =ymean - m*xmean
b
```

PREDICT

```python
def predict(value):
    predict = m*value + b 
    return predict
```
